```python
df["Date"] = pd.to_datetime(df["Date"],format='%d/%m/%Y') #Dátummá alakítjuk a Date-t
```

```python
df=df.set_index(["Date"]) #Date-ból indexet képzünk
```

# Ábrázoljuk az idősort

```python
plt.figure().set_figwidth(12)
plt.xlabel("Dátum")
plt.ylabel("Esetek száma")
plt.title("Bárányhimlős esetek alakulása Magyarországon 2005 és 2015 között")
plt.plot(df.index, df["ÖSSZES_ESET"], color="blue")
plt.show()
```

# A magasan kiugró 2008-as értéket, és az alacsonyan kiugró 2014 év körüli értékeket interpolációval kicseréljük.

```python
df.iloc[[np.where(df["ÖSSZES_ESET"]>2500)]]=np.nan  
df.iloc[[np.where((df["ÖSSZES_ESET"]<50) & (df.index.strftime("%Y")=="2013"))]]=np.nan  
df.iloc[[np.where((df["ÖSSZES_ESET"]<100) & (df.index.strftime("%Y")=="2014") & (df.index.strftime("%m")=="04"))]]=np.nan 

df["ÖSSZES_ESET"]=df["ÖSSZES_ESET"].interpolate()

plt.figure().set_figwidth(12)
plt.xlabel("Dátum")
plt.ylabel("Esetek száma")
plt.title("Bárányhimlős esetek alakulása Magyarországon 2005 és 2015 között")
plt.plot(df.index, df["ÖSSZES_ESET"], color="blue")
plt.show()
```

# Ábrázoljuk a mozgó átlagot, illetve szórás, és megnézzük változik-e az időben. 

```python
rolmean = df.rolling(window=52).mean() #gördülő átlag egy évnyi ráhagyással
rolstd = df.rolling(window=52).std() #gördülő szórás egy évnyi ráhagyással
plt.figure().set_figwidth(12)
plt.xlabel("Dátum")
plt.ylabel("Esetek száma")
plt.title("Bárányhimlős esetek alakulása Magyarországon 2005 és 2015 között")
plt.plot(df.index, df["ÖSSZES_ESET"], color="blue", label="Eredeti")
plt.plot(df.index, rolmean, color='red', label="Gördülő átlag")
plt.plot(df.index, rolstd, color='green', label="Gördülő szórás")
plt.legend(loc='best')
plt.show()
```

# Transzformációt végzünk, hogy csökkentsük a trendhatást. Vesszük az idősor logaritmusát.

```python
df_log = np.log(df)
moving_avg = df_log.rolling(52).mean() #mozgóátlag egy évnyi ráhagyással
plt.figure().set_figwidth(12)
plt.plot(df_log)
plt.plot(moving_avg, color='red')
```

```python
#képezzük a különbséget
df_log_moving_avg_diff = df_log - moving_avg
#kidobjuk a NaN értékeket
df_log_moving_avg_diff.dropna(inplace=True)
#ábrázoljuk a kapott eredményt
rolmean = df_log_moving_avg_diff.rolling(window=52).mean() #gördülő átlag egy évnyi ráhagyással
rolstd = df_log_moving_avg_diff.rolling(window=52).std() #gördülő szórás egy évnyi ráhagyással
plt.figure().set_figwidth(12)
plt.xlabel("Dátum")
plt.title("Bárányhimlős esetek alakulása Magyarországon 2005 és 2015 között")
plt.plot(df_log_moving_avg_diff, color="blue", label="Logaritmikus")
plt.plot(rolmean, color='red', label="Gördülő átlag")
plt.plot(rolstd, color='green', label="Gördülő szórás")
plt.legend(loc='best')
plt.show()
```

# Szétválasztjuk az idősorunkat trend, szezonális, és reziduális értékekre.

```python
from statsmodels.tsa.seasonal import seasonal_decompose
decomposition = seasonal_decompose(df_log)  

trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

plt.subplot(411)
plt.ylabel("Eredeti")
plt.plot(df_log) 

plt.subplot(412)
plt.ylabel("Trend")
plt.plot(trend)

plt.subplot(413)
plt.ylabel("Szezonalitás")
plt.plot(seasonal)

plt.subplot(414)
plt.ylabel("Reziduális")
plt.plot(residual)
plt.tight_layout()
```

# Leválasztjuk a trendet és a szezonalitást

```python
# A trendet és a szezonalitást leválasztják az adatoktól, és modellezzük a maradványokat. 
df_log_decompose = residual
df_log_decompose.dropna(inplace=True)

rolmean = df_log_decompose.rolling(window=52).mean() #gördülő átlag egy évnyi ráhagyással
rolstd = df_log_decompose.rolling(window=52).std() #gördülő szórás egy évnyi ráhagyással

plt.figure().set_figwidth(12)
plt.xlabel("Dátum")
plt.title("Bárányhimlős esetek alakulása Magyarországon 2005 és 2015 között")
plt.plot(df_log_decompose, color="blue", label="Maradvány")
plt.plot(rolmean, color='red', label="Gördülő átlag")
plt.plot(rolstd, color='green', label="Gördülő szórás")
plt.legend(loc='best')
plt.show()
```

# Elvégezzük a Dickey-Fuller tesztet, hogy megbizonyosodjunk arról, hogy az idősorunk stacionárius-e.

```python
from statsmodels.tsa.stattools import adfuller
print ("A Dickey-Fuller teszt eredménye:")
dftest = adfuller(df_log_decompose, autolag = 'AIC')
dfoutput = pd.Series(dftest[0:4], index= ['Test Statistic', 'p-value', '#Lags Used', 'Number of Observations Used'])
    
for key, value in dftest[4].items():
    dfoutput['Critical Value (%s) ' %key] = value
print(dfoutput)
```

A test statictic értéke kisebb, mint a kritikus értékeink, és a p-érték is kisebb jóval, mint 0,05. Így a H0 hipotézist nem fogadjuk el, az idősorunk stacionárius.

```python
arima_model = SARIMAX(df_log, order = (1,1,2), seasonal_order = (1,0,1,12))
result_SARIMA = arima_model.fit()
plt.figure().set_figwidth(12)
plt.plot(df_log)
plt.plot(result_SARIMA.fittedvalues[1:], color='red')
plt.title('RSS: %.4f'%sum((result_SARIMA.fittedvalues - df_log["ÖSSZES_ESET"])**2))
```

```python
predict=result_SARIMA.predict()
plt.figure().set_figwidth(12)
plt.plot(predict[1:])
plt.plot(df_log)
```

# A végeredmény grafikusan...

```python
beteg = result_SARIMA.forecast(steps=20) #20 hétnyi előrejelzés
plt.figure().set_figwidth(12)
plt.plot(predict)
plt.plot(beteg)
```
